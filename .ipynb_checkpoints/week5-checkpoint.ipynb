### Geographical data download

```python
lat_long = []

def geo(zip_code, API_URL):
    zip = str(zip_code)
    response = urllib.request.urlopen(API_URL + zip)
    xml_result = response.read()
    response.close()
    elem = xml.etree.ElementTree.fromstring(xml_result)
    long = elem.findtext(".//x")
    lat = elem.findtext(".//y")
    data = [zip_code, lat, long]
    return data

URL = 'http://geoapi.heartrails.com/api/xml?method=searchByPostal&postal='

for i in range(len(df)):
    lat_long.append(geo(df.Postcode[i], URL))

df_lat_long = pd.DataFrame(lat_long, columns=['Postcode', 'Latitude', 'Longitude'])  
df_lat_long.head()
```

```python
df = pd.merge(df, df_lat_long, on='Postcode')
df.head()
```

```python
df = df[['Postcode', 'Area', 'Population', 'Latitude', 'Longitude']]
df.head()
```

### Data preporcessing to create the regression model

As a preprocessing of data first extract urban area by population.  
Then get restaurant numbers of areas from Foursquare API.  
This time I set the searching diameter as 6km for each areas which is around 5min by car.  
(Most of Toyota-city do not have train so basically people move by car or bus.)

```python
df['Population'] = df['Population'].str.replace(',','')
df['Population'] = df['Population'].astype(int)
df.head()
```

```python
#Extract urban area by population filtering.
df_ooi = df[df['Population'] > 2000].reset_index()
df_ooi = df_ooi[['Postcode', 'Area', 'Population', 'Latitude', 'Longitude']]
print(len(df_ooi))
df_ooi.head()
```

```python
#Visualization
address  =  'Toyota'
geolocator = Nominatim(user_agent="Toyota_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude

Toyota_map = folium.Map(location=[latitude, longitude], zoom_start=11)

for lat, lng, population, postcode, area in zip(df_ooi['Latitude'], df_ooi['Longitude'], df_ooi['Population'], df_ooi['Postcode'], df_ooi['Area']):
    population=population/150
    if population == 0:
        break
    else:
        #print(population)
        label = '{}, {}'.format(postcode, area)
        label = folium.Popup(label, parse_html=True)
        folium.CircleMarker(
            [lat, lng],
            radius=population,
            popup=label,
            color='red',
            fill=True,
            fill_color='pink',
            fill_opacity=0.7,
            parse_html=False).add_to(Toyota_map)

    
Toyota_map
```

```python
#Get the number of restaurans from Foursquare API.
CLIENT_ID = 'GB5DI5'# your Foursquare ID
CLIENT_SECRET = 'R5K5ZLHYIISG' # your Foursquare Secret
VERSION = '20180604'
LIMIT = 100
query_list = ['食事', 'レストラン', '飲食', 'ラーメン' '料理', 'パスタ', 'カフェ', 'フード']
radius = 6000
num_of_res = []
res_num = 0
```

```python
for i in range(len(df_ooi)):
    latitude = df_ooi.Latitude[i]
    longitude = df_ooi.Longitude[i]

    for query in query_list:
        url = 'https://api.foursquare.com/v2/venues/search?client_id={}&client_secret={}&ll={},{}&v={}&query={}&radius={}&limit={}'.format(CLIENT_ID, CLIENT_SECRET, latitude, longitude, VERSION, query, radius, LIMIT)
        results = requests.get(url).json()["response"]['venues']
        
        if len(results) == 0:
            break
        else:
            res_num = res_num + len(results)

    num_of_res.append(res_num)
    res_num = 0
    #print(num_of_res)
```

```python
df_ooi['ResNum'] = pd.DataFrame(num_of_res)
df_ooi.head(10)
```

### Linear model creation

Next I will create the Linear model from preprocessec data by using sklearn.  
First visualize the population vs restaurant number.  
Then create the mode.

```python
import matplotlib.pyplot as plt
```

```python
x = df_ooi.Population
y = df_ooi.ResNum
plt.scatter(x, y, s=20, c=None, marker='o')
```

```python
from sklearn import linear_model
LM = linear_model.LinearRegression()
x_train = np.asanyarray(df_ooi[['Population']])
y_train = np.asanyarray(df_ooi[['ResNum']])
LM.fit(x_train, y_train)
```
